```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

summary_df = df_3[['client_id', 'qty', 'shipping_price', 'line_price', 'profit']].copy()
print(summary_df)
#suggested by Arti AI at work to organize by client_id
summary_df = summary_df.groupby('client_id')[['qty', 'shipping_price', 'line_price', 'profit']].sum()
summary_df.head(20)

#reset the client_id back into a column as groupby made it an index.
summary_df = summary_df.reset_index()
summary_df = summary_df[summary_df['client_id'].isin(top_5_clients)]
summary_df.head()

#set the client_id as the index for the dataframe.
#filtering_columns_df = summary_df.set_index('client_id', inplace=True)
#filtering_columns_df
```
